# Object Oriented Parking Garage
<h5>Your parking garage class should have the following methods:
<h5>takeTicket()
<ul><li>This should decrease the amount of tickets available by 1
<li>This should decrease the amount of parkingSpaces available by 1</ul>
<h5>paysForParking()
<ul><li>Display an input that waits for an amount from the user and store it in a variable
<li>If the payment variable is not empty then (meaning the ticket has been paid -> display a message to the user that their ticket has been paid and they have 15mins to leave
<li>This should update the "currentTicket" dictionary key "paid" to True</ul><h5>leaveGarage()
<ul><li>If the ticket has been paid, display a message of "Thank You, have a nice day"
<li>-If the ticket has not been paid, display an input prompt for payment
<li>Once paid, display message "Thank you, have a nice day!"
<li>Update parkingSpaces list to increase by 1 (meaning add to the parkingSpaces list)
<li>Update tickets list to increase by 1 (meaning add to the tickets list)</ul>
<h4>You will need a few attributes as well:</h4>
<ul><li>tickets -> list
<li>parkingSpaces -> list
<li>currentTicket -> dictionary</ul>



```python
class ParkingGarage:
    def __init__(self):
        self.tickets = [1] * 10
        self.parkingSpaces = [1] * 10
        self.currentTicket = {}

    def takeTicket(self):
        print(f'Tickets Available: {len(self.tickets)}')
        print(f'Parking Spaces Available: {len(self.parkingSpaces)}')
        if len(self.tickets) > 0 and len(self.parkingSpaces) > 0:
            self.tickets.pop()
            self.parkingSpaces.pop()
            self.currentTicket = {"paid": False}
            print("Here's your ticket: [PARKING--TICKET]")
        else:
            print('No tickets or parking spaces available!')
            return False

    def payForParking(self):
        if self.currentTicket:
            payment = input("Please enter your payment amount: ")
            if payment:
                self.currentTicket["paid"] = True
                print("Your ticket has been paid and you have 15mins to leave.")
            else:
                print("Please enter your payment amount.")
        else:
            print("You do not have any tickets to pay.")

    def leaveGarage(self):
        print('Checking if ticket was payed...')
        if self.currentTicket.get("paid"):
            print("Thank You, have a nice day!")
            self.parkingSpaces.append(1)
            self.tickets.append(1)
        else:
            print("Please pay for your parking ticket.")
            self.payForParking()
            self.leaveGarage()

garage = ParkingGarage()

garage.takeTicket()
print("=====================================")
garage.payForParking()
print("=====================================")
garage.leaveGarage()

print("=====================================")
print("=====================================")

garage.takeTicket()
print("=====================================")
garage.takeTicket()
print("=====================================")
garage.leaveGarage()
print("=====================================")
garage.payForParking()
print("=====================================")
garage.leaveGarage()
```
